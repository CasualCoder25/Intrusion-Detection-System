```python
X_10_90=np.array(X_10_90,dtype=float)
Y_10_90=np.array(Y_10_90,dtype=float)
```

```python
X_50_50=np.array(X_50_50,dtype=float)
Y_50_50=np.array(Y_50_50,dtype=float)
```

```python
X_90_10=np.array(X_90_10,dtype=float)
Y_90_10=np.array(Y_90_10,dtype=float)
```

```python
X_10_90_train,X_,Y_10_90_train,Y_=train_test_split(X_10_90,Y_10_90,test_size=0.4,random_state=1)
X_10_90_cv,X_10_90_test,Y_10_90_cv,Y_10_90_test=train_test_split(X_,Y_,test_size=0.5,random_state=1)
print(X_10_90_train.shape,Y_10_90_train.shape)
print(X_10_90_cv.shape,Y_10_90_cv.shape)
print(X_10_90_test.shape,Y_10_90_test.shape)
```

```python
X_50_50_train,X_,Y_50_50_train,Y_=train_test_split(X_50_50,Y_50_50,test_size=0.4,random_state=1)
X_50_50_cv,X_50_50_test,Y_50_50_cv,Y_50_50_test=train_test_split(X_,Y_,test_size=0.5,random_state=1)
print(X_50_50_train.shape,Y_50_50_train.shape)
print(X_50_50_cv.shape,Y_50_50_cv.shape)
print(X_50_50_test.shape,Y_50_50_test.shape)
```

```python
X_90_10_train,X_,Y_90_10_train,Y_=train_test_split(X_90_10,Y_90_10,test_size=0.4,random_state=1)
X_90_10_cv,X_90_10_test,Y_90_10_cv,Y_90_10_test=train_test_split(X_,Y_,test_size=0.5,random_state=1)
print(X_90_10_train.shape,Y_90_10_train.shape)
print(X_90_10_cv.shape,Y_90_10_cv.shape)
print(X_90_10_test.shape,Y_90_10_test.shape)
```

```python
scaler_10_90=StandardScaler()
scaler_10_90.fit_transform(X_10_90_train)
X_10_90_train_scaled=scaler_10_90.transform(X_10_90_train)
X_10_90_cv_scaled=scaler_10_90.transform(X_10_90_cv)
X_10_90_test_scaled=scaler_10_90.transform(X_10_90_test)
```

```python
scaler_50_50=StandardScaler()
scaler_50_50.fit_transform(X_50_50_train)
X_50_50_train_scaled=scaler_50_50.transform(X_50_50_train)
X_50_50_cv_scaled=scaler_50_50.transform(X_50_50_cv)
X_50_50_test_scaled=scaler_50_50.transform(X_50_50_test)
```

```python
scaler_90_10=StandardScaler()
scaler_90_10.fit_transform(X_90_10_train)
X_90_10_train_scaled=scaler_90_10.transform(X_90_10_train)
X_90_10_cv_scaled=scaler_90_10.transform(X_90_10_cv)
X_90_10_test_scaled=scaler_90_10.transform(X_90_10_test)
```

```python
metrics_list_10_90=[]
```

```python
models_10_90=[]
```

```python
lambdas_=[0.000001]
```

```python
learning_rate=[0.0001]
```

```python
recall=tf.keras.metrics.Recall()
precision=tf.keras.metrics.Precision()
```

```python
for lambda_ in lambdas_:
    temp1=[]
    models1=[]
    for lr in learning_rate:
        print(f'Training model lambda={lambda_} learning_rate={lr}...')
        temp2=[]
        model=tf.keras.models.Sequential([
            tf.keras.layers.Dense(units=512,activation='relu',kernel_regularizer=tf.keras.regularizers.L2(lambda_)),
            tf.keras.layers.Dense(units=512,activation='relu',kernel_regularizer=tf.keras.regularizers.L2(lambda_)),
            tf.keras.layers.Dense(units=128,activation='relu',kernel_regularizer=tf.keras.regularizers.L2(lambda_)),
            tf.keras.layers.Dense(units=1,activation='sigmoid',kernel_regularizer=tf.keras.regularizers.L2(lambda_))
        ])

        model.compile(
            loss=tf.keras.losses.BinaryCrossentropy(),
            optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
            metrics=['accuracy',recall,precision]
        )

        tf.random.set_seed(1)
        model.fit(X_10_90_train_scaled,Y_10_90_train,epochs=3)

        metrics=model.evaluate(X_10_90_train_scaled,Y_10_90_train)
        temp2.append(metrics)

        metrics=model.evaluate(X_10_90_cv_scaled,Y_10_90_cv)
        temp2.append(metrics)

        metrics=model.evaluate(X_10_90_test_scaled,Y_10_90_test)
        temp2.append(metrics)

        models1.append(model)
        temp1.append(temp2)
        print('Training Done')

    metrics_list_10_90.append(temp1)
    models_10_90.append(models1)
```
